```python
from extract import extractor

# levels.json generated with `[...document.querySelectorAll('.level_icon')].map((e) => e.level)`
levels = extractor.load_from_extract('./extract/levels.json')
levels
```

```python
import time
import subprocess

runtimes = {}

for k, v in levels.items():
    # Daily
    if isinstance(k, str):
        continue

    map_str = '\n'.join(v['grid'])

    begin = time.time()
    subprocess.run([f'solvers/{v['width']}', str(v['par'])], input=map_str, text=True)
    end = time.time()

    runtimes[k] = end - begin
```

```python
sum(runtimes.values()), max(runtimes.values())
```
